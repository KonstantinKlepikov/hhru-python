```python
import pandas as pd
import ast
from collections import Counter
```

```python
data = pd.read_csv('../data/result_deep_sorted_11-16-2022(15-37).csv')
data.info()
```

```python
count = Counter()

for row in data['key_skills']:
    skill = ast.literal_eval(row)
    skill = [n.strip().lower() for n in skill]
    for s in skill:
        count[s] += 1
```

```python
count.most_common()
```
